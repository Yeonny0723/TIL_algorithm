1. K-Goodness String

```python
from collections import deque

T = int(input())

for i in range(T):
    N, K = map(int, input().split())
    q = deque(input())
    count = 0
    while len(q) > 1:
        if q.popleft() != q.pop():
            count +=1
            
    print(f"Case #{i+1}:", abs(K-count))
```

2. L Shaped Plots

- All the coordinates with value 1 can be an edge of L shape
- Brute force all coords to see the max strectch
- Calculate the number of L shape can be located within max strech for each coord

```python
# T = int(input())
R, C = map(int, input().split())
matrix = [list(map(int, input().split())) for _ in range(4)]
```

```python
def max_stretch(m,x,y,a,b,res=0):
    '''
    m: matrix
    x,y: coordinates
    '''
    print(x+a, y+b)
    if m[x+a][y+b]:
        res += 1
        max_stretch(m,x+a,y+b,a,b,res)
    else:
        return res
```
